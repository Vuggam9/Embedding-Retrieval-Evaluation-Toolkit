# Retrieval Evaluation Walkthrough

This notebook walks through the evaluation toolkit step by step.

Run the cells in order to load documents, chunk them, embed them, retrieve results, and inspect benchmark outputs.

```python
import sys
from pathlib import Path

sys.path.append(str(Path.cwd().parent / 'src'))
```

```python
from dataset import load_sample_documents, load_sample_queries
from chunking import chunk_by_words

documents = load_sample_documents()
queries = load_sample_queries()
chunks = chunk_by_words(documents, chunk_size=20, overlap=5)

len(documents), len(queries), len(chunks)
```

```python
from embeddings import load_embedding_model, embed_texts

model = load_embedding_model('sentence-transformers/all-MiniLM-L6-v2')
chunk_embeddings = embed_texts(model, [chunk.text for chunk in chunks])
len(chunk_embeddings), len(chunk_embeddings[0])
```

```python
from retrieval import build_faiss_index, search_index

index = build_faiss_index(chunk_embeddings)
query_embedding = embed_texts(model, [queries[0].query])[0]
results = search_index(index, query_embedding, chunks, top_k=3)
[(chunk.doc_id, chunk.text) for chunk in results]
```
